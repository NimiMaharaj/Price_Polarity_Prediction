```python
import pandas as pd
from datetime import datetime
```

```python
df = pd.read_excel('../data/Training File.xlsx')
```

```python
df
```

```python
df.columns
```

```python
df['date_str'] = df['Date'].dt.strftime('%d-%m-%Y')
df['start_time'] = df['Time Block'].str.split(' - ').str[0]
df['datetime'] = pd.to_datetime(
    df['date_str'] + ' ' + df['start_time'],
    dayfirst=True
)
df = df.set_index('datetime').sort_index()
```

```python
df
```

```python
df['G DAM MCP'].plot(figsize=(12,4))
```

```python
# xlimit and ylimit
df['G DAM MCP'].plot( 
    xlim =[ '2025-08-01', '2025-08-01 23:45:00'],  
    figsize=(12,4))
```
